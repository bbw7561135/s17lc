```python
import rsnr
import pylab
```

```python
x=rsnr.do_one()
```

```python
print(x)
```

```python
pylab.figure(1,(6,6))
pylab.loglog(x['Age'],2*x['R'])
pylab.xlabel('Age (yrs)')
pylab.ylabel(('D (pc)'))
```

```python
pylab.figure(1,(6,6))
pylab.loglog(x['Age'],x['v'])
pylab.xlabel('Age (yrs)')
pylab.ylabel((r'v (km s$^{-1}$)'))
```

```python
d=815*3.08e21
dfactor=4*3.14*d*d

pylab.figure(1,(6,6))
pylab.loglog(2*x['R'],x['L_r']/dfactor)
pylab.xlabel('D (pc)')
pylab.ylabel((r'L$_R$(ergs cm$^{-2}$ s$^{-1}$ Hz${-1}$)'))
```
